```python
import pandas as pd
from scipy.stats import spearmanr

df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar14/healthy_breast.tsv', index_col = 0, sep = '\t')
df['spearman cor'] = [spearmanr(df.loc['SPI1'], df.loc[gene])[0] for gene in df.index]
df['p-value'] = [spearmanr(df.loc['SPI1'], df.loc[gene])[1] for gene in df.index]

df = df.loc[(df['spearman cor'].abs() > 0.8) & (df['p-value'] < 0.05)]
```
