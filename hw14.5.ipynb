```python
import pandas as pd
from sklearn.manifold import TSNE

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar14/human_coronavirus_aln_scores.tsv', index_col = 0, sep = '\t')
df = df**(-1) #чем больше score тем меньше будет расстояние
# большой score это хорошо, маленькое расстояние - тоже круто
```

```python
model = TSNE
df[['tsne-2d-one', 'tsne-2d-two']] = TSNE(n_components = 2, perplexity = 30, metric = 'precomputed').fit_transform(df) #default perplexity is 30
```

perplexity = 30

```python
plt.tight_layout()
sns.scatterplot(x = 'tsne-2d-one', y = 'tsne-2d-two', data = df, palette = 'Set3')
```
